```python
import time
import winsound
def myalarm():
    try:
        s=list(map(int,input().split()))
        print(s)
        t_sec=s[0]*60*60+s[1]*60+s[2]
        time.sleep(t_sec)
        frequency=2500
        duration=5000 #1000ms=1sec
        winsound.Beep(frequency,duration)
    except:
        pass
myalarm()
```
